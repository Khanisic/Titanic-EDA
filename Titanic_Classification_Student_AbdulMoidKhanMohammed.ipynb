# Classification Models For Titanic

Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
eda=pd.read_csv('Cleaned_Titanic.csv')
```

```python
eda.head()
```

```python
eda.drop(['Unnamed: 0'],axis='columns',inplace=True) 
```

```python
eda.head()
```

```python
y = eda.iloc[:,:1].values
y
```

```python
X = eda.iloc[:, 1:14].values
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =
0.25, random_state = 0)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

## Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
print("Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,2),
"%")
```

### Accuracy: 78.03 %

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200, n_jobs= 2, random_state
= 0)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
y_test
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print("Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,2),
"%")
```

## Accuracy : 78.92 %

# K Neighbour Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,2),
"%")
```

## Accuracy : 74.44 %

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
# Representing LinearRegression as lr(Creating LinearRegression Object)
lr = LogisticRegression()

# fit the model to the training data
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
confusion = confusion_matrix(y_test, y_pred)
print(confusion)
```

```python
TN = confusion [0,0]
FP = confusion [0,1]
FN = confusion [1,0]
TP = confusion [1,1]
```

```python
print(confusion)
print ("TN: ", TN)
print ("FP: ", FP)
print ("FN: ", FN)
print ("TP: ", TP)
```

```python
confusion_matrix = pd.DataFrame(confusion)
confusion_matrix.columns = ['Predicted No Subscription', 'Predicted Yes Subscription']
confusion_matrix = confusion_matrix.rename(index = {0 : 'Actual No Subscription', 1 : 'Actual Yes Subscription'})
confusion_matrix
```
